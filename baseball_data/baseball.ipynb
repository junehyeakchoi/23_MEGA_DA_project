```python
#선수를 입력하면 그 선수의 기록이 나오는것
seasons = 18.0
name = "손아섭"

season = df[df["시즌"] == seasons]
seasonsplayer = season[season["선수명"] == name]
seasonsplayer
```

```python
# n 시즌에 경기당 안타 기댓값
a = seasonsplayer["경기"]
b = seasonsplayer["홈런"]
v_a = a.values.tolist()[0]
v_b = b.values.tolist()[0]

value = v_b / v_a
value
```
